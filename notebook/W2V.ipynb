```python
import gensim, logging
from tqdm import tqdm
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
sentences = []
```

```python
fopen = open("../corpus/clean_text.txt","r").readlines()
```

```python
for i in tqdm(range(len(fopen))):
    sentence = fopen[i].split()
    sentences.append(sentence)
```

```python
# train word2vec on the two sentences
model = gensim.models.Word2Vec(sentences, size=300, min_count=3, window=5, negative=5, iter=100)
```

```python
print(model)
```

```python
model.wv.most_similar(positive=['setnov'], topn=10)
```

```python
model.wv.doesnt_match("jokowi pergi rapat".split())
```

```python
print(model.predict_output_word(['presiden', 'indonesia', 'joko']))
```
